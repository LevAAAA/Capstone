Problem - find most overcrowded schools in Toronto neighborhoods

First step - Get data from wikipedia - list of Secondary schools in Toronto

```python
!wget -q -O 'wiki.html' https://en.wikipedia.org/wiki/List_of_secondary_schools_in_the_Toronto_District_School_Board
print('Data downloaded!')
```

Get only list of schools

```python
import pandas as pd
tables = pd.read_html('wiki.html', header=0)
tables = tables[0]
tables.head()
```

```python
headings = ['Name', 'Location', 'Population', 'EQAO Academic Math score', 'ESL population', 'OSSLT pass', 'Special programs', 'Picture']
tables.columns = headings
tables.head()
```

Get only required columns

```python
dt_toronto = tables[['Name', 'Location', 'Population', 'EQAO Academic Math score', 'OSSLT pass', 'Special programs']]

dt_toronto.head()
```

To get addresses of scholls used TDSB site and some conversion done from html to csv

```python
#!wget -q -O 'tdsb.html' https://www.tdsb.on.ca/Find-your/School/Secondary

!wget -q -O 'schools.csv' https://raw.githubusercontent.com/LevAAAA/Capstone/master/schools.csv
    
print('Data downloaded!')
```

```python
tb = pd.read_csv('schools.csv')

tb.head()
```

```python
len(tb)
len(dt_toronto)
```

Removed extra characters from Names of school

```python
dt_toronto['Name']=dt_toronto['Name'].str.replace('.', '')
dt_toronto

```

Joined dataframes from Wikipedia with scores and from TDSB with addresses

```python
df_schools=pd.merge(dt_toronto,
            tb,
            left_on='Name',
            right_on='School',
            how='left')
df_schools.head()
```

Done clean-up for column names and added column PC with 3 first letters from postal code

```python
df_schools.columns = ['Name', 'Location', 'Population', 'EQAO Academic Math score',
       'OSSLT pass', 'Special programs', 'School', 'Grades', 'Address',
       'City', 'Prov', 'PostalCode', 'Phone', 'PC']

df_schools['PC']=df_schools['PostalCode'].str.slice(1,4)
df_schools
```

```python
df_schools['PC']=df_schools['PC'].astype(str)
```

Get location data based on Postal Codes

```python
!wget -q -O 'Geospatial_data.csv' https://cocl.us/Geospatial_data
print('Data downloaded!')
```

```python
geo = pd.read_csv('Geospatial_data.csv', header=0)
geo.head()
```

```python
geo.shape
```

Added coordinate columns for each school

```python
df_schools_geo=df_schools.join(geo.set_index('Postal Code'), on='PC')
df_schools_geo
```

```python
print('The dataframe has {} Schools in {} Locations.'.format(
        df_schools_geo.shape[0],
        len(df_schools_geo['Location'].unique())
    )
)
```
