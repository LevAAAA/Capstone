Problem - find most overcrowded schools in Toronto neighborhoods

First step - Get data from wikipedia - list of Secondary schools in Toronto

```python
!wget -q -O 'wiki.html' https://en.wikipedia.org/wiki/List_of_secondary_schools_in_the_Toronto_District_School_Board
print('Data downloaded!')
```

Get only list of schools

```python
import pandas as pd
tables = pd.read_html('wiki.html', header=0)
tables = tables[0]
tables.head()
```

```python
headings = ['Name', 'Location', 'Population', 'EQAO Academic Math score', 'ESL population', 'OSSLT pass', 'Special programs', 'Picture']
tables.columns = headings
tables.head()
```

Get only required columns

```python
dt_toronto = tables[['Name', 'Location', 'Population', 'EQAO Academic Math score', 'OSSLT pass', 'Special programs']]

dt_toronto.head()
```

To get addresses of scholls used TDSB site and some conversion done from html to csv

```python
#!wget -q -O 'tdsb.html' https://www.tdsb.on.ca/Find-your/School/Secondary

!wget -q -O 'schools.csv' https://raw.githubusercontent.com/LevAAAA/Capstone/master/schools.csv
    
print('Data downloaded!')
```

```python
tb = pd.read_csv('schools.csv')

tb.head()
```

```python
len(tb)
len(dt_toronto)
```

Removed extra characters from Names of school

```python
dt_toronto['Name']=dt_toronto['Name'].str.replace('.', '')
dt_toronto

```

Joined dataframes from Wikipedia with scores and from TDSB with addresses

```python
df_schools=pd.merge(dt_toronto,
            tb,
            left_on='Name',
            right_on='School',
            how='left')
df_schools.head()
```

Done clean-up for column names and added column PC with 3 first letters from postal code

```python
df_schools['PC']=""
```

```python
df_schools.columns = ['Name', 'Location', 'Population', 'EQAO Academic Math score',
       'OSSLT pass', 'Special programs', 'School', 'Grades', 'Address',
       'City', 'Prov', 'PostalCode', 'Phone', 'PC']

df_schools['PC']=df_schools['PostalCode'].str.slice(1,4)
df_schools
```

Get location data based on Postal Codes

```python
!wget -q -O 'Geospatial_data.csv' https://cocl.us/Geospatial_data
print('Data downloaded!')
```

```python
geo = pd.read_csv('Geospatial_data.csv', header=0)
geo.head()
```

```python
geo.shape
```

Added coordinate columns for each school

```python
df_schools_geo=df_schools.join(geo.set_index('Postal Code'), on='PC')
df_schools_geo
```

```python
print('The dataframe has {} Schools in {} Locations.'.format(
        df_schools_geo.shape[0],
        len(df_schools_geo['Location'].unique())
    )
)
```

Imported libraries for maps

```python
import numpy as np # library to handle data in a vectorized manner

import json # library to handle JSON files

!conda install -c conda-forge geopy --yes # uncomment this line if you haven't completed the Foursquare API lab
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans

!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
import folium # map rendering library

print('Libraries imported.')
```

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="to_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))

#latitude = 43.653963
#longitude = -79.387207
```

Created map of school with raitings

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, school, eqao, osslt in zip(df_schools_geo['Latitude'], 
                                        df_schools_geo['Longitude'], 
                                        df_schools_geo['Location'], 
                                        df_schools_geo['Name'], 
                                        df_schools_geo['EQAO Academic Math score'],
                                        df_schools_geo['OSSLT pass']):
    label = '{}, {} with math EQAO {} OSSLT {}'.format(borough, school, eqao, osslt)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

Groupped school population based on Postal code

```python
df_pc=df_schools_geo.groupby(['PC'], as_index = False)['Population'].agg('sum')
#df_pc.columns=['PC','Population']
```

```python
df_pc
```

Created clasters based on population on Postal code - excluding empty postal codes

```python
df_pc.loc[df_pc['Population'] < 500, 'Cluster Labels'] = 0
df_pc.loc[df_pc['Population'] >= 500, 'Cluster Labels'] = 1
df_pc.loc[df_pc['Population'] >= 1000, 'Cluster Labels'] = 2
df_pc.loc[df_pc['Population'] >= 1500, 'Cluster Labels'] = 3
df_pc.loc[df_pc['Population'] >= 2000, 'Cluster Labels'] = 4
df_pc.loc[df_pc['Population'] >= 3000, 'Cluster Labels'] = 5

df_pc = df_pc.astype({'Cluster Labels': int, 'Population': int})

df_pc
```

```python
df_pc.dtypes
```

Created full list of Postal codes for Toronto and school population based on postal code

```python
df_pc_geo=df_pc.join(geo.set_index('Postal Code'), on='PC')
df_pc_geo_full=pd.merge(geo, df_pc, left_on='Postal Code',right_on='PC', how='outer')
#df_pc_geo

df_pc_geo_full['PC']=df_pc_geo_full['Postal Code']
df_pc_geo_full
```

Replaced NaN for 0

```python
df_pc_geo_full['Population']=df_pc_geo_full['Population'].fillna(0)
df_pc_geo_full['Cluster Labels']=df_pc_geo_full['Cluster Labels'].fillna(0)

df_pc_geo_full = df_pc_geo_full.astype({'Cluster Labels': int, 'Population': int})

df_pc_geo_full
```

Created clasters for school based on popupalion in Postal Code

```python
df_pc_geo_full.loc[df_pc_geo_full['Population'] == 0, 'Cluster Labels'] = 0
df_pc_geo_full.loc[df_pc_geo_full['Population'] > 0, 'Cluster Labels'] = 1
df_pc_geo_full.loc[df_pc_geo_full['Population'] >= 500, 'Cluster Labels'] = 2
df_pc_geo_full.loc[df_pc_geo_full['Population'] >= 1000, 'Cluster Labels'] = 3
df_pc_geo_full.loc[df_pc_geo_full['Population'] >= 2000, 'Cluster Labels'] = 4
df_pc_geo_full.loc[df_pc_geo_full['Population'] >= 3000, 'Cluster Labels'] = 5

df_pc_geo_full
```

Created map with all Postal codes and number of student population

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=10)

kclusters = 6

# set color scheme for the clusters
#x = np.arange(kclusters)
#ys = [i + x + (i*x)**2 for i in range(kclusters)]
#colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
#rainbow = [colors.rgb2hex(i) for i in colors_array]
rainbow = ['white','blue','green','yellow','orange','red']

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster, population in zip(df_pc_geo_full['Latitude'], df_pc_geo_full['Longitude'], df_pc_geo_full['PC'], df_pc_geo_full['Cluster Labels'], df_pc_geo_full['Population']):
    label = folium.Popup(str(poi) + ' Population ' + str(population), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster],
        fill=True,
        fill_color=rainbow[cluster],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```

White circles - postal codes without public schools in area

Yellow, Orange and Red - most overcrowded areas
